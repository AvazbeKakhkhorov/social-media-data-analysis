## Step 1: Importing Required Libraries

As the name suggests, the first step is to import all the necessary libraries that will be used in the project. In this case, we need pandas, numpy, matplotlib, seaborn, and random libraries.

Pandas is a library used for data manipulation and analysis. Numpy is a library used for numerical computations. Matplotlib is a library used for data visualization. Seaborn is a library used for statistical data visualization. Random is a library used to generate random numbers.

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Define the list of categories
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
categories
```

```python
# Step 3: Set number of entries
n = 500  # or any number you want
```

```python
# Step 4: Generate random data
data = {
    'Date': pd.date_range('2021-01-01', periods=n),
    'Category': [random.choice(categories) for _ in range(n)],
    'Likes': np.random.randint(0, 10000, size=n),
    'Retweets': np.random.randint(0, 5000, size=n)  # Added Retweets too for you!
}
data
```

```python
# Step 5: Create a DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# 4. Print the count of each Category
print("\n🔹 Count of each Category:")
df['Category'].value_counts()
```

## Cleaning the data


```python
# 1. Remove all null (missing) values
df = df.dropna()
```

```python
# 2. Remove duplicate rows
df = df.drop_duplicates()
```

```python
# 3. Convert the 'Date' column to datetime format (important for time-based analysis)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# 4. Convert the 'Likes' column to integer type (although it's already int, but in real data it's a good practice)
df['Likes'] = df['Likes'].astype(int)
```

```python
# To ensure 'Retweets' is integer
df['Retweets'] = df['Retweets'].astype(int)
```

```python
# 5. Print to check everything is clean
print("🔹 Cleaned DataFrame Head:")
df.head()
```

```python
print("\n🔹 DataFrame Info After Cleaning:")
print(df.info())
```

## #5 Visualize and Analyze the data

```python
# 1. Histogram of Likes
plt.figure(figsize=(10,6))
sns.distplot(df['Likes'], kde=True, color='blue', bins=30, hist=True)
plt.title('Distribution of Likes')
plt.xlabel('Number of Likes')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 2. Boxplot of Likes by Category
plt.figure(figsize=(12,8))
sns.boxplot(x='Category', y='Likes', data=df, palette='Set2')
plt.title('Likes Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Number of Likes')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# 3. Statistics

# Mean of 'Likes' overall
overall_mean_likes = df['Likes'].mean()
print(f"🔹 Overall Mean of Likes: {overall_mean_likes:.2f}")
```

```python
# Mean of 'Likes' per 'Category'
mean_likes_per_category = df.groupby('Category')['Likes'].mean()
print("\n🔹 Mean of Likes per Category:")
mean_likes_per_category
```
